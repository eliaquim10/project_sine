```python
import pandas as pd
import numpy as np
import os
```

```python
pontuas = {
    "'C\"" : "'C'",
    "'AB\"" : "'AB'",
    "'A\"" : "'A'",
    "'B\"": "'B'",
    "'A/B\"" : "'A/B'",
    "AUTOMOTIVA\"" : "AUTOMOTIVA'",
    "'B\"" : "'B'",
    "'B\"" : "'B'",
    "MASCULINO\"" : "MASCULINO'",
    "FEMININO\"" : "FEMININO'",
    "INFORMÁTICA\"": "INFORMÁTICA'",
    "EM geral'\"": "EM geral'",
    # "CONDUÍTES);" 
    # "DESPACHAR;" 
    # "ORGANIZAR;"
    #  "ESTOQUE;" 
    #  "PREÇOS;"
    # "série\";":
  }
```

```python

def corrige_file(fileName):
  # leitura
  entrou = False
  
  s = ""
  with open(fileName, mode="r", encoding="ISO8859") as arquivo:
    for linha in arquivo:        
      # print(linha)
      for c in range(len(linha)-1):
        if (linha[c] == '"' ):
          if(entrou == False):
            entrou = True
            s += linha[c]
          elif(linha[c+1] == ";"):
            entrou = False
            s += linha[c]
            for key, value in pontuas.items():
              if (s[c-len(key):c] == key):
                entrou = True
                s = s[:-len(key)] + value
          elif(linha[c+1] == "\n"):
            entrou = False
            s += linha[c]
          else:
            s += "\'"
        elif(entrou):
          if (linha[c] == ";"):
            # print("tratou", ":")
            s += ":"
          elif(linha[c] == "\n"):
            s += " "
          else:
            s += linha[c]
        else:
          if (linha[c] =="\n"):
            print(linha[c-1], linha[c], linha[c+1], "->")
          s += linha[c]
        
      s += linha[-1]
  
  # escrita
  with open("copy_" + fileName, "w", encoding="UTF8") as arquivo:
    # open(fileName[:-4] + "copy" + fileName[-4:], "w")
    arquivo.write(s)
    arquivo.flush()
    arquivo.close()    
```

```python

# dirs = ["D_ETL_IMO_EXTRACAO_SINE_ABERTO_VAGAS_PR.csv"]
def reader(fileName):
  return pd.read_csv(fileName, sep=";", encoding="UTF8", 
  error_bad_lines=False,
  # names=["CODIGO_OCUPACAO", "TITULO_OCUPACAO", "OBSERVACOES_OCUPACAO", "CODIGO_MUNICIPIO_IBGE", "CEP", "VALOR_SALARIO", "QTD_VAGAS", "TIPO_CONTRATACAO", "QTD_EXPERIENCIA", "ESCOLARIDADE", "REQUERIDO_DEFICIENCIA", "DEFICIENCIAS"]
  )
```

```python
list_dataset = []
n_linhas = 0
for dir in dirs:
  if (".csv" in dir):
    print(dir[-6:-4])
    corrige_file("dataset/vagas/" + dir)
    dataset = reader("copy_dataset/vagas/" + dir)
    n_linhas +=len(dataset)
    dataset["ESTADO"] = dir[-6:-4]
    list_dataset.append(dataset.copy())

# dataset = pd.read_csv("dataset\\trabalhadores\D_ETL_IMO_EXTRACAO_SINE_ABERTO_TRABALHADORES_AC.csv", delimiter=";", encoding="ISO8859")

len(list_dataset)
all_dataset = pd.concat(list_dataset)
all_dataset.shape
```

```python
def remove_espaco(qtd):
  def elements(palavra):
    palavra = palavra
    if(type(palavra)==str and "-" in palavra):
      palavra = " - ".join(palavra.split("-")[-qtd:])
    return palavra
  return elements
# dataset["ESCOLARIDADE"]
all_dataset["ESCOLARIDADE"] = all_dataset["ESCOLARIDADE"].apply(remove_espaco(1))
all_dataset["TIPO_CONTRATACAO"] = all_dataset["TIPO_CONTRATACAO"].apply(remove_espaco(1))
# dataset["DEFICIENCIAS"] = dataset["DEFICIENCIAS"].apply(remove_espaco(2))
```

```python
all_dataset = all_dataset[all_dataset["REQUERIDO_DEFICIENCIA"] != "REQUERIDO_DEFICIENCIA" ]
all_dataset.head(5)
# .cat.categories
```

```python
# remove dados com alguma irreguralidade
def correcao_def(s):
  return (s==1 or s==0 or s=="1" or s=="0")
def correcao_int(s):
  try:
    int(s)
    return True
  except:
    return False
all_dataset = all_dataset[all_dataset["REQUERIDO_DEFICIENCIA"].apply(correcao_def)]
all_dataset = all_dataset[all_dataset["QTD_VAGAS"].apply(correcao_int)]

all_dataset.head(5)

```

```python
all_dataset["ESCOLARIDADE"] = all_dataset["ESCOLARIDADE"].astype("category")
all_dataset["TIPO_CONTRATACAO"] = all_dataset["TIPO_CONTRATACAO"].astype("category")
all_dataset["TITULO_OCUPACAO"] = all_dataset["TITULO_OCUPACAO"].astype("category")
all_dataset["ESTADO"] = all_dataset["ESTADO"].astype("category")
all_dataset["REQUERIDO_DEFICIENCIA"] = all_dataset["REQUERIDO_DEFICIENCIA"].astype("int")
all_dataset["QTD_EXPERIENCIA"] = all_dataset["QTD_EXPERIENCIA"].astype("int")
all_dataset["QTD_VAGAS"] = all_dataset["QTD_VAGAS"].astype("int")
# all_dataset["DEFICIENCIAS"] = all_dataset["DEFICIENCIAS"].astype("category")
```
